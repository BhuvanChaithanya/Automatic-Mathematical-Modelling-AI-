```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#for basic understanding i will be only using 2 data points
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_trian = {y_train}")
```

```python
#number of training examples is denoted by variable m and in this case is 2
```

```python
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples id {m}")
```

```python
m = len(x_train)
print(f"Number of training examples is {m}")
```

```python
#training examples x_i, y_i

i = 0

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
#scatter is used to plot all the points

plt.scatter(x_train, y_train, marker = 'x',c= 'r')
#setting the title

plt.title("housing prices")

#setting y and x axis label

plt.ylabel("prince in 1000s of dollars")

plt.xlabel("size in 1000s of sqft")

plt.show()
```

```python
#model function

w = 200
b = 100

print(f"w = {w}")
print(f"b = {b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i]+b
        
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train,w, b)

plt.plot(x_train, tmp_f_wb, c ='b',label = 'Prediction')

plt.scatter(x_train, y_train, marker = 'x', c = 'r', label= ' Actual Values')

plt.title('housing prices')

plt.ylabel('price in 1000s of dollars')

plt.xlabel('size in 1000s of sqft')
plt.legend()


plt.show()
```
